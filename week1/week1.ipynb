```python
def factors(n) : 
    factorList = []
    for i in range(1, n+1) : 
        if n%i == 0 : 
            factorList.append(i)

    return factorList
```

```python
def prime(n) : 
    return (factors(n) == [1, n])

def prime2(n) : 
    return (len(factors(n)) == 2)
```

```python
primeList = []

for i in range(1, 101) : 
    if prime(i) : 
        primeList.append(i)

print(primeList) 
```

```python
def primeBetter(n) : 
    for i in range(2, int(pow(n, 0.5))+1) : 
        if (n%i) == 0 : 
            return False
    return True   
```

```python
def firstPrimes(m) : 
    count, i, primeList = 0, 2, []
    while count < m : 
        if primeBetter(i) : 
            primeList.append(i)
            count += 1
        i += 1 
    return primeList

firstPrimes(10)
```

```python
def primeDiffs(n) : 
    lastPrime = 2
    primeDiffDict = {}

    for i in range(3, n+1) : 
        if primeBetter(i) : 
            d = i - lastPrime
            lastPrime = i 
            if d in primeDiffDict.keys() : 
                primeDiffDict[d] = primeDiffDict[d] + 1
            else : 
                primeDiffDict[d] = 1 
    return primeDiffDict

primeDiffs(10)
```
